```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
g = 9.8
m = .15
b = .015
vi = v = 25
yi = y = 300
t = 0
dt = 0.001
time = [0]
height = [yi]
velocity = [vi]
```

```python
def f():
    global t , v , y
    t = t +dt
    y = y + v*dt
    v = v +(-g-(b/m)*v)*dt
```

```python
while True:
    f()
    if y<=0:
        break
    time.append(t)
    height.append(y)
    velocity.append(v)
```

```python
time_axis = np.array(time)
height_axis = np.array(height)
velocity_axis = np.array(velocity)

figure, axis = plt.subplots(1,2)
axis[0].plot(time_axis, height_axis)
axis[0].set_title('Height vs Time')
axis[1].plot(time_axis, velocity_axis)
axis[1].set_title('Velocity vs Time')
plt.show()
```

```python
terminal_velocity = -m*g/b
final_velocity = velocity_axis[-1]
flight_time = time_axis[-1]
ratio = final_velocity/terminal_velocity

print("Flight time=", flight_time.round(2),'s')
print("Final Velocity=",final_velocity.round(2),'m/s')
print("ratio=",ratio.round(2))
```
